## Face Detection => Crop and Resize to train model

```python
import cv2
```

```python
faceCascade = cv2.CascadeClassifier('../data/haarcascade_frontalface_default.xml')
```

```python
happy_face = cv2.imread('../data/happy_face.jpeg',0)
```

```python
import matplotlib.pyplot as plt
plt.imshow(cv2.cvtColor(happy_face,cv2.COLOR_BGR2RGB))
```

```python
faces = faceCascade.detectMultiScale(happy_face)
```

```python
happy_face.shape
```

```python
faces
```

```python
len(faces)
```

```python
for x,y,w,h in faces:
    cv2.rectangle(happy_face,(x,y),(x+w,y+h),(0,255,0),2)
    plt.imshow(cv2.cvtColor(happy_face[y:y+h,x:x+w],cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
plt.imshow(cv2.cvtColor(happy_face,cv2.COLOR_BGR2RGB))
```

```python
cropImage = happy_face[y:y+h,x:x+w]
```

```python
plt.imshow(cv2.cvtColor(cropImage,cv2.COLOR_BGR2RGB))
```

## Two or more Faces Condition

```python
multipleImg = cv2.imread("../data/people-dining-outdoors-at-summer-party.webp",0)
```

```python
faceCascade2 = cv2.CascadeClassifier('../data/haarcascade_frontalface_default.xml')
mulFace = faceCascade2.detectMultiScale(multipleImg)
```

```python
plt.imshow(cv2.cvtColor(multipleImg,cv2.COLOR_BGR2RGB))
```

```python
len(mulFace)
```

```python
for x,y,w,h in mulFace:
    temp = multipleImg[y:y+h,x:x+w]
    detectTemp = faceCascade2.detectMultiScale(temp)
    if len(detectTemp) == 0:
        print(-999)
    else:
        for nx,ny,nw,nh in detectTemp:
            plt.imshow(cv2.cvtColor(multipleImg[y:y+h,x:x+w],cv2.COLOR_BGR2RGB))
            plt.show()
```
