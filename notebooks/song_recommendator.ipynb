```python
from mpl_toolkits.mplot3d import Axes3D

from sklearn.cluster import KMeans
```

```python
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(data[numer_cols])
    Sum_of_squared_distances.append(km.inertia_)
```

```python
plt.plot(K,Sum_of_squared_distances)
```

```python
numClusters = data['mood'].nunique()
```

```python
numClusters
```

```python
from sklearn.metrics import silhouette_score
for n_clusters in range(2,15):
    kModel = KMeans(n_clusters=n_clusters,algorithm="elkan")
    preds = kModel.fit_predict(data[numer_cols])
    centers = kModel.cluster_centers_

    score = silhouette_score (data[numer_cols], preds, metric='euclidean')
    print ("For n_clusters = {}, silhouette score is {})".format(n_clusters, score))
```

```python
from sklearn.decomposition import PCA

kmeans = KMeans(n_clusters=numClusters,algorithm="elkan")
kmeans = kmeans.fit(data[numer_cols])
y_kmeans = kmeans.predict(data[numer_cols])
pca = PCA(n_components=4)
principal_components = pca.fit_transform(data[numer_cols])
```

```python
pc = pd.DataFrame(principal_components)
pc['label'] = y_kmeans
pc.columns = ['w','v','x', 'y','label']

#plot data with seaborn
cluster = sns.lmplot(data=pc, x='w', y='v', hue='label', 
                   fit_reg=False, legend=True, legend_out=True)
```
