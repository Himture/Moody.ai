```python
import pandas as pd
```

```python
def filterGenre(data):
    if data[0]=='[':
        data = data.strip('[]').replace(' ','').replace("'",'')
        return data.split(',')
    else:
        return list(data)
```

```python
data = pd.read_csv('../data/AnimeWorld.csv')
```

```python
data.Genre = data.Genre.apply(filterGenre)
```
